```python
# calculate and display statistical summary of petal_length for each species separately.
def stats_summry(data):
    return{
        'Mean':data.mean(),
        'STD dev': data.std(),
        '25th percentile':data.quantile(0.25),
        '50th percentile':data.quantile(0.5),
        '75th percentile':data.quantile(0.75)
    }
for sp in df2['species'].unique(): # → gets unique species names
    print('\n stats for',sp)
    print(stats_summry(df2[df2['species']==sp]['petal_length']))
```
