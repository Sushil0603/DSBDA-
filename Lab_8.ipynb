```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df.isnull().sum()
```

```python
sns.countplot(x='survived', data=df) # Creates a count plot (bar chart)
plt.title("survival Count") # 0 → Not survived 1 - survived
plt.show()
```

```python
sns.countplot(x='sex' ,hue='survived' ,data=df) # hue for color
plt.title("Survival by Gender")
plt.show()
```

```python
sns.countplot(x='pclass' , hue='survived' , data=df)
plt.title("Survival by Passenger Class")
plt.show() 
# part B: Histogram of Fare Distribution
```

```python
plt.hist(df['fare'].dropna(), bins=20 , color = 'orange')
plt.title("Distribution of Fare")
plt.xlabel("Fare")
plt.ylabel("frequency")
plt.show()
```
