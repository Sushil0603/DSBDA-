```python
sns.boxplot(x='Sex' , y='Age' , hue ='Survived' , data=df)
plt.title("Age Distribution by Gender and Survival")
plt.show()
```

```python
# 1) Male Survival & Age: For males, the median age of survivors (orange box) is lower than those who did not survive (blue box). 
# The survivor box also extends much lower toward age 0.

# 2) Female Survival & Age: For females, the median age for survivors is actually slightly higher than for non-survivors. 
# The boxes are more similar in height, but the survivor box (orange) is shifted slightly upward.

# 3) Outliers: There are several outliers (the circles) in the male category, 
# particularly one very elderly male survivor around age 80 and several elderly males who did not survive (ages 65–75).

# 4) Overall Spread: The interquartile range (the middle 50% of the data represented by the boxes) 
# for all groups generally falls between 20 and 40 years old.
```
